```python
#!/usr/bin/env bash
# Batch convert all .mod files in a directory to .point files using IMOD's model2point

set -euo pipefail

# Directory containing .mod files; default = coords_original
MOD_DIR="${1:-coords_original}"

# Make sure the directory exists
if [[ ! -d "$MOD_DIR" ]]; then
  echo "Error: directory '$MOD_DIR' does not exist."
  exit 1
fi

echo "Converting .mod files in: $MOD_DIR"

# Loop over all .mod files
shopt -s nullglob
for mod_file in "$MOD_DIR"/*.mod; do
  base_name="$(basename "$mod_file" .mod)"
  out_file="$MOD_DIR/${base_name}.point"

  echo "  -> $mod_file  -->  $out_file"

  # Call IMOD's model2point
  # -float: output coordinates in floating point format
  # You can add more options here if needed (e.g., -scattered, -contours, etc.)
  model2point -float "$mod_file" "$out_file"
done

echo "Done."
```
